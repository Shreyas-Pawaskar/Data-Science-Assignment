##  ANN with Hyperparameter Tuning

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report
```

```python
df.columns
```

```python
X = df.drop(columns=["letter"])
X
```

```python
y = df["letter"]
y
```

```python
# Encode target labels

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
print(y)
```

```python
# Normalize features

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

# Build ANN model

```python
def create_model(hidden_units=64, activation='relu', optimizer='adam'):
    model = keras.Sequential([
        keras.layers.Dense(hidden_units, activation=activation, input_shape=(X_train.shape[1],)),
        keras.layers.Dense(32, activation='relu'),
        keras.layers.Dense(len(np.unique(y)), activation='softmax')  
    ])
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Train model
model = create_model()
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

## # Evaluate model

```python
y_pred = np.argmax(model.predict(X_test), axis=1)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

```python
# Hyperparameter tuning 
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV

model_wrapper = KerasClassifier(model=create_model, hidden_units=64, activation='relu', optimizer='adam', verbose=0)

# Define parameter grid (prefix 'model__' to pass correctly)
param_grid = {
    'model__hidden_units': [32, 64, 128],
    'model__activation': ['relu', 'tanh'],
    'model__optimizer': ['adam', 'sgd'],
    'batch_size': [16, 32, 64]
}

# Use GridSearchCV for hyperparameter tuning
grid = GridSearchCV(model_wrapper, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
# Best parameters and accuracy

print("Best parameters:", grid.best_params_)
print("Best accuracy:", grid.best_score_)
```
