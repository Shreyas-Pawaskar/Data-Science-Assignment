```python
from sklearn.preprocessing import StandardScaler

df = df.drop(columns=['ID#'])
```

```python
df.head()
```

```python
scale = StandardScaler()
scale_df = scale.fit_transform(df)
```

```python
# Convert scaled data back to DataFrame

df = pd.DataFrame(scale_df, columns=df.columns)
```

```python
df.head()
```

```python
# Plot the distribution of key numerical features
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
features = ["Balance", "Bonus_miles", "Flight_miles_12mo", "Days_since_enroll", "Award?", "Bonus_trans"]

for i, feature in enumerate(features):
    row, col = divmod(i, 3)
    sns.histplot(df[feature], bins=30, kde=True, ax=axes[row, col])
    axes[row, col].set_title(f"Distribution of {feature}")

plt.tight_layout()
plt.show()
```

```python
# applying clustering using K-Means, Hierarchical Clustering, and DBSCAN

from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA
```

```python
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df["Cluster_KMeans"] = kmeans.fit_predict(df)
```

```python
# Use a subset of the data for Hierarchical Clustering (random 1000 samples)
subset_data = df.sample(n=1000, random_state=42)

# Hierarchical Clustering 

hierarchical = AgglomerativeClustering(n_clusters=optimal_k, linkage="ward")
subset_data["Cluster_Hierarchical"] = hierarchical.fit_predict(subset_data)


#  DBSCAN Clustering 

dbscan = DBSCAN(eps=1.2, min_samples=5)  
df["Cluster_DBSCAN"] = dbscan.fit_predict(df)


# Reduce dimensions using PCA 

pca_full = PCA(n_components=2)
pca_result_full = pca_full.fit_transform(df.drop(columns=["Cluster_KMeans", "Cluster_DBSCAN"]))


# PCA for the hierarchical subset

pca_subset = PCA(n_components=2)
pca_result_subset = pca_subset.fit_transform(subset_data.drop(columns=["Cluster_Hierarchical"]))

# Visualize Hierarchical Clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_result_subset[:, 0], y=pca_result_subset[:, 1], hue=subset_data["Cluster_Hierarchical"], palette="viridis", alpha=0.7)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Hierarchical Clustering Visualization (PCA Reduced)")
plt.legend(title="Cluster")
plt.show()

# Visualize DBSCAN Clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_result_full[:, 0], y=pca_result_full[:, 1], hue=df["Cluster_DBSCAN"], palette="coolwarm", alpha=0.7)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("DBSCAN Clustering Visualization (PCA Reduced)")
plt.legend(title="Cluster")
plt.show()
```

# Cluster Analysis and Insights
### K-Means (K=5)
- Identified 5 well-separated clusters.

- Customers were grouped based on account balance, miles traveled, and bonus transactions.

- Silhouette score helped confirm the optimal number of clusters.

- Suitable when the number of clusters is known and data is well-structured.

---

### Hierarchical Clustering
- Performed on a subset of 1000 data points due to memory constraints.

- Clusters are more interpretable since dendrograms can be used.

- The separation of clusters is similar to K-Means but slightly more flexible.

- 

### DBSCAN (Density-Based)
Identified dense clusters but also detected noise points (outliers).

More suitable for irregularly shaped clusters compared to K-Means.

Less effective when clusters have varying densities.
